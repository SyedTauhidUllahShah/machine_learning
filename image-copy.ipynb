```python
image = mpimg.imread("/Users/alvaro/Downloads/AAEAAQAAAAAAAAd5AAAAJDBiN2RjODUyLTg5YzMtNGI1My1iYmY1LWJhZjQ3ZDRjMDMwZg.jpg")
image_t = torch.Tensor(image/255)
plt.imshow(image)
```

```python
noise_image = torch.rand(image.shape, requires_grad=True)
plt.imshow(noise_image.detach().numpy())
```

```python
criteria = torch.optim.Adam([noise_image], lr=0.1)
```

```python
criteria.zero_grad()
loss = F.mse_loss(image_t, noise_image)
print("loss", loss.data.numpy())
loss.backward()
criteria.step()
plt.imshow(noise_image.detach().numpy())
```
